```python
def normalisation(A,Vmax=255,Vmin=0): #permettre de faire en sorte que tout les pixels soit representer
    return np.round(((Vmax-Vmin)/(A.max()-A.min()))*(A-A.min())+Vmin) 
```

```python
plt.imshow(normalisation(mo),cmap='gray');
```

```python
def egalisation(img,Vmin=0,Vmax=255):
    x,n=np.unique(img,return_counts=True)
    N=np.cumsum(n)
    B= ((Vmax-Vmin)/img.size)*np.interp(img,x,N)+Vmin;
    return B.astype(int)


def egalisationimg(img,Vmin=0,Vmax=255):
    IMG=img.copy()
    COUCHE=np.zeros((img.shape[0],img.shape[1],img.shape[2]))
    k=img.shape[2]
    for j in range(k):
        COUCHE[:,:,j]=egalisation(img[:,:,j])
        IMG[:,:,j]=COUCHE[:,:,j]
    return IMG
```

```python
plt.imshow(egalisationimg(cheval))
```
